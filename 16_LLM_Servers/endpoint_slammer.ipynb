Now, let's SLAM IT.

```python
import asyncio

async def send_request(llm, idx):
    try:
        response = await llm.ainvoke(
            f"How much wood could a wood chuck chuck if a wood chuck could chuck wood? (Request {idx})"
        )
        print(f"Response {idx}: {response.content[:60]}...")
    except Exception as e:
        print(f"Request {idx} failed: {e}")

async def main():
    tasks = [send_request(llm, i) for i in range(1, 5)]
    await asyncio.gather(*tasks)

await main()
```

## General Deworming Schedule for Cats

| Age / Risk Group | Typical Interval | Notes |
|------------------|------------------|-------|
| **Kittens (0–12 weeks)** | Every **2–3 weeks** | Most kittens are born with a few roundworms; frequent deworming keeps the worm burden low. |
| **Kittens (12–24 weeks)** | **Monthly** | Worm burden usually drops, but monthly deworming is still recommended until 6 months of age. |
| **Adult cats (≥ 6 months)** | **Every 3–6 months** | A quarterly deworming (four times a year) is common for indoor cats. |
| **Outdoor or hunting cats** | **Every 2–3 months** | Higher exposure to infected prey or soil → more frequent treatment. |
| **Cats with a history of parasites** | **As advised by your vet** | Some cats may need a more aggressive schedule. |

### Why the Schedule Matters

- **Kittens** are especially vulnerable to roundworms, which can cause vomiting, diarrhea, and weight loss.  
- **Adult cats** can still harbor tapeworms, hookworms, or roundworms, especially if they hunt or roam.  
- **Outdoor exposure** increases the risk of infection, so a shorter interval is prudent.

### Practical Tips

1. **Ask your vet:** They’ll tailor the schedule to your cat’s lifestyle and local parasite prevalence.  
2. **Use a single, broad-spectrum dewormer** (e.g., fenbendazole, pyrantel pamoate, or a combination product) unless a specific parasite is suspected.  
3. **Keep a record** of deworming dates and the product used.  
4. **Watch for signs:** diarrhea, weight loss, scooting, or visible worms in the stool — these warrant a vet visit and possibly a stool exam.

```python
from langgraph_sdk import get_sync_client



client = get_sync_client(url="http://localhost:2024")
for chunk in client.runs.stream(
    None,  # Threadless run
    "simple_agent",  # Assistant id from langgraph.json (assistants)
    input={
        "messages": [
            {
                "role": "human",
                "content": "How often should I deworm my cat?",
            }
        ]
    },
    stream_mode="updates",
):
    print(f"Receiving new event of type: {chunk.event}...")
    print(chunk.data)
    print("\n\n")
```
